```python
import matplotlib.pyplot as plt
```

```python
sales = [500, 340, 270]
products = ['Phone', 'Other Electronics', 'Furnitures']
```

```python
plt.pie(sales, labels = products);
```

```python
plt.pie(sales, [0.1,0,0], labels = products, autopct = '%2.1f'); # 0.1 = 10% away from center, autopct = significant figure
```
