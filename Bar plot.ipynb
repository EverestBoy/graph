# Object oriented plotting

```python
rcParams['figure.figsize'] = 4,3
fig  = plt.figure();
ax = fig.add_axes([0,0,1,1]);
ax.set_xticks([1,3,5,7,10])
ax.set_yticks([min(y), np.mean(y), 8.5, max(y)])
ax.plot(x,y)
```
