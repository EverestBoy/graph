# Line plot

It is very useful in finding trend.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,3,2,4,5,7,6,8,9]
```

```python
plt.plot(x, y, color = '#172b4d')
```

```python
plt.plot(x, y, color = '#172b4d', marker = 'o', mew = 5);
```

```python
plt.plot(x, y, color = '#172b4d', marker = 'o', mew = 5, ls = ':');
```

```python
plt.plot(x, y, color = '#172b4d', marker = 'o', mew = 5, ls = '-.');
```
